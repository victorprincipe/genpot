# Creating a Baselined General-Purpose Potential for Molecular Crystals

Hello, welcome to what should be an extensive notebook on how to create a _baselined_ general-purpose machine learning potential for molecular crystals. Everything should be commented relatively well above each section and in the code cells, but if there is anything unclear, feel free to email me at victorprincipe1206@gmail.com.

The potential that will be created will be baselined on DFTB energies and forces, and will target PBE-DFT level of theory with Grimme-D2 dispersion correcion. The entire dataset contains 2238 crystals, with 101 configurations per crystal (1 geomoetry-optimised and 100 finite-temperature configurations). These configurations were obtained from short, 500 fs simulations using i-PI as the simulation engine and Quantum Espresso for energy calculations. Please contact Edgar Engel for more information on these simulations. 

The notebook is split up into 11 main sections:
1. Initial sorting of configurations in each crystal using farthest-point sampling, in order to obtain the 11 most diverse configurations in each crystal (11 x 2238 = 24,618 configurations)
2. Calculating DFTB energies and forces for the ~24.6k configurations, in order to calculate the delta between PBE-D2 (the high level of theory) and DFTB (the baseline potential)
3. Create a validation set and _initial_ training set, by simply removing 1000 random structures from the ~24.6k configurations
4. Create a simple Sklearn baselined energy model with the per-atom binding energy for each element
5. Use hybrid PCovFPS to sort the ~24.6k structures according to structural and energetic diversity
6. Optimise SOAP hyperparameters by training Sklearn KRR models only on energies
7. Use hybrid PCovFPS for feature selection, reducing the size of the final feature vector
8. Create learning curves using various numbers of features and samples, in order to determine optimal number and see whether the model is saturating
9. Create the final model using optimal paramters as determined from the above steps
10. Test the model on the validation set structures, and compare results to predictions made using Aditi's general-purpose baselined model
11. Test the model on the CSD-1k test set structures, and compare results to predictions made using Aditi's general-purpose baselined model
    - Note that the CSD-1k test set contains 500 crystals that are not in the training set, with 2 configurations per crystal (1 geometry-optimised and 1 finite-temperature configuration)

Please note that this model is created using Librascal (`rascal`), which can be installed using the instructions found at https://github.com/lab-cosmo/librascal. The notebook also requires Scikit-matter (`skmatter`), with installation instructions found at https://github.com/lab-cosmo/scikit-matter. Other required packages include `numpy`, `ase`, `pandas`, `matplotlib`, `tqdm`, `time`, and `pickle`.

```python
#Define useful functions
def do_fps(x, d=0):
    """
    Function for performing farthest-point-sampling for a given feature matrix.
    d gives the number of farthest-point-sampled feature vectors that will be outputted. 
    If d==0, the entire set will be FPS-sorted.
    Returns the FPS-sorted IDs, as well as the FPS distances.
    """
    if d == 0 : d = len(x)
    n = len(x)
    iy = np.zeros(d, int)
    # faster evaluation of Euclidean distance
    n2 = np.sum(x**2,axis=1)
    iy[0] = 0
    dl = n2 + n2[iy[0]] - 2* np.dot(x, x[iy[0]])
    dss = []
    for i in range(1,d):
        iy[i] = np.argmax(dl)
        nd = n2 + n2[iy[i]] - 2*np.dot(x,x[iy[i]])
        dl = np.minimum(dl, nd)
        dss.append(max(dl))
    return iy,dss

def get_forces(frames):
    frc = np.zeros((np.sum([len(frm) for frm in frames]),3))
    iat = 0
    for ifrm,frm in enumerate(frames):
        frc[iat:iat+len(frm)] = frm.get_array('Delta_forces') 
        iat += len(frm)
    return frc

def get_features(atoms_objects, batch_size=1000, progress_bar=True):
    '''Get average feature vectors for each structure in a list of atoms objects. 
        This requires soap to be set.
    '''
    struct_feat=[] #create empty feature vector of n_structures x n_features
    got_frame = False
    for frm in atoms_objects:
        if got_frame == False: 
            if all(sp in frm.get_atomic_numbers() for sp in [1,6,7,8]) == True:
                all_species = frm
                got_frame = True
    for i in tqdm(range(int(len(atoms_objects)/batch_size+0.9999)), leave=False,
                  disable = not progress_bar): #Get feature vectors in batches of 1000
        for frm in atoms_objects[i*batch_size:(i+1)*batch_size]:
            frm.wrap(eps=1e-13) #wrap atoms in unit cell
        frames=[all_species.copy()] #add initial frame with all species
        frames.extend(atoms_objects[i*batch_size:(i+1)*batch_size]) #extend initial frame with frames of batch
        manager = soap.transform(frames) #calculate soap features for all structures in batch
        env_feat_batch = manager.get_features(soap)[len(all_species):] #get feature vectors for all structures in batch
        # (having removed the initial frame)
        atom_counter=0 #count atoms
        for ifrm,frm in enumerate(atoms_objects[i*batch_size:(i+1)*batch_size]): #iterate over frames in batch
            nat=len(frm) #count atoms in frame
            struct_feat.append(np.mean(env_feat_batch[atom_counter:atom_counter+nat],axis=0)) #average vectors
            # for atoms in the same frame. Done by averaging vectors between atom_counter and atom_counter + nat
            atom_counter+=nat #add atoms to counter
    struct_feat_array=np.asarray(struct_feat)
    return struct_feat_array
```

# FPS-sort configurations of each crystal
Here we will FPS-sort the 101 configurations (1 geometry-optimised + 100 MD) of each crystal. The 11 most diverse (out of 101) will be taken from each crystal and added to a new database.

```python
#Define soap hyperparameters for FPS sorting configurations - use cheap values for this
hypers = dict(soap_type="PowerSpectrum",
              interaction_cutoff=4, #cutoff distance in angstroms
              max_radial=6, #no. of radial basis functions
              max_angular=4, #no. of angular basis functions
              gaussian_sigma_constant=0.4, #sigma width (i.e. amount of 'smearing')
              gaussian_sigma_type="Constant",
              cutoff_function_type="RadialScaling",
              cutoff_smooth_width=0.5,
              cutoff_function_parameters=
                    dict(
                            rate=1,
                            scale=2,
                            exponent=2
                        ),
              radial_basis="GTO",
              normalize=True,
              optimization=
                    dict(
                            Spline=dict(
                               accuracy=1.0e-05
                            )
                        ),
              compute_gradients=False #don't care about forces right now
              )
soap = SphericalInvariants(**hypers)
```

```python
#FPS-sort initial training set to get 11 configurations per crystal
db_FPS = {} #new dictionary of 11 most diverse configurations of each crystal
selector = FPS(n_to_select=11, initialize=0) #initialize at 0 so that first selection is always geop

for name in tqdm(names): #loop over all crystals
    all_feats = []
    for ifrm, frm in enumerate(db[name]): #loop over each configuration of the given crystal
        frm.wrap(eps=1e-13)
        feats = soap.transform(frm).get_features(soap) #get average feature vector of structure
        all_feats.append(np.mean(feats, axis=0))
    selector.fit(all_feats) #FPS sort feature vectors of all structures of the crystal
    frames = []
    for i in selector.selected_idx_: #append IDs of 11 most diverse (FPS-sorted) frames to new database
        frames.append(db[name][i])
    db_FPS[name] = frames
```

```python
#Rename "forces" array to "PBE-D2_forces" to avoid confusion, and rename "energy" to "PBE-D2_energy"
for name in names:
    for frm in db[name]:
        frm.arrays['PBE-D2_forces'] = frm.arrays['forces'].copy() #add new array with good name
        frm.set_array('forces', None) #remove old array
        frm.info['PBE-D2_energy'] = frm.info['energy'].copy() #add new info with energy
        frm.info.pop('energy') #remove old energy info
```

```python
#Compute DFTB energies and forces, and add them to the frames
for name in tqdm(names, desc="Crystal Progress", leave=False):
    for frm in tqdm(db[name], desc="Config Progress", leave=False):
        
        kpts = np.array2string(frm.info['kpts'], separator=' ')[1:-1] #Get K-points
        if len(kpts) != 5: #Print k points if they are longer than useful (for security) 
            print(name, frm) 
        kpts += ' 1.0' #add the k points "weight" for DFTB+
        
        calc=Dftb(**calc_preliminary.todict()) #Get preliminary calculator
        calc.set(Hamiltonian_KPointsAndWeights="{"+kpts+"}") #Set k points
        
        #Set Hubbard derivs and max angular momenta for the relevant species present
        if "H" in frm.get_chemical_symbols():
            calc.set(Hamiltonian_HubbardDerivs_H=-0.1857)
            calc.set(Hamiltonian_MaxAngularMomentum_H='"s"')
        if "C" in frm.get_chemical_symbols():
            calc.set(Hamiltonian_HubbardDerivs_C=-0.1492)
            calc.set(Hamiltonian_MaxAngularMomentum_C='"p"')
        if "N" in frm.get_chemical_symbols():
            calc.set(Hamiltonian_HubbardDerivs_N=-0.1535)
            calc.set(Hamiltonian_MaxAngularMomentum_N='"p"')
        if "O" in frm.get_chemical_symbols():
            calc.set(Hamiltonian_HubbardDerivs_O=-0.1575)
            calc.set(Hamiltonian_MaxAngularMomentum_O='"p"')
        
        frm.set_calculator(calc) #set calculator for the frame
        
        frm.info['DFTB_energy'] = frm.get_potential_energy() #calculate potential energy, and add it as info
        frm.arrays['DFTB_forces'] = frm.get_forces() #calculate forces, and add them as info
```

# PCovFPS sample selection
We will now PCovFPS sort the _initial_ training set.

Please note that PCovFPS will augment the distance measurement between samples with a term that includes the distance in property space as well. It also uses a mixing parameter that, essentially, linearly interpolates between maximising distance in feature space and property space. Here, our _property_ is of course the baselined per-atom energies in our initial "training" set.

Please see https://arxiv.org/pdf/2012.12253.pdf for more information on PCovFPS.

```python
#Create baselined per-atom binding energies of training and validation sets
yred_train = y_train/np.sum(X_train,axis=1)-mlr.predict(X_train)/np.sum(X_train,axis=1)
yred_val = y_val/np.sum(X_val,axis=1)-mlr.predict(X_val)/np.sum(X_val,axis=1) 
```

# Small Hyperparam Optimisation

We will now optimise SOAP hyperparameters for generation of feature vectors. Please note some non-exhaustive limitations of this approach as implemented:
- We are optimising using only energies as the property to train on, not forces as we will do when actually training the GAP model
- We are optimising using the structure-average feature vector, not the atomic feature vectors as will be used when training the GAP model on forces too
- We are using Scikit-Learn's KRR model here, not Librascal
- We are using the full feature vector, rather than the reduced vector after PCovFPS feature selection

Two loops were created below, but the first one can almost be ignored because I did not optimise the `gaussian_sigma_constant`. Also, although the first loop suggests hypers such as `interaction_cutoff` of 8 Angstrom, `max_radial` of 12 and `max_angular` of 8, these hypers are too high if you want to maintain some computational efficiency. Please look at the second loop for the most _applicable_ results.

```python
#Loop for SOAP hyperparameter optimisation
n_train = 10000

results=[]
count=0; 

#Iterate over interaction cuttoff
for rc in [4.,6.,8.]:
    hypers['interaction_cutoff']=rc
    
    #Iterate over nmax
    for n in [6,9,12]:
        hypers['max_radial'] = n
        
        #Iterate over lmax
        for l in [4,6,8]:
            start = time()
            hypers['max_angular'] = l
            soap = SphericalInvariants(**hypers) #redefine soap with new parameters
            #print("Iteration: ", count+1, "/ 36") #keep count of iteration
            train_feat=get_features(train_frames[:n_train], batch_size=250) #get features for training set
            val_feat=get_features(val_frames, batch_size=250) #get features for validation set
            count+=1
            
            #Normalise feature vectors
            for feat in train_feat:
                feat/=np.linalg.norm(feat)
            for feat in val_feat:
                feat/=np.linalg.norm(feat)

            #Iterate over zeta values
            for zeta in [2.,3.]:
                K = np.dot(train_feat,train_feat.T)**zeta #compute covariance matrix for training
                Kval = np.dot(val_feat,train_feat.T)**zeta #compute covariance matrix for validation

                #Iterate over regularization values
                minerr = 1e24 ; minreg = 0
                for reg in np.logspace(-9,-5,5):
                    KRR = KernelRidge(kernel='precomputed',alpha = reg)
                    KRR.fit(K,yred_train[:n_train]) #Fit KRR model
                    err = np.std(yred_val - KRR.predict(Kval)) #Get error (std. dev) of vaidation set
                    if err < minerr: #only get reg corresponding to minimum error
                        minerr = err.copy()
                        minreg = reg.copy()
                results.append([minerr, rc, n, l, zeta, minreg]) #append data to results
            print(f"Results using rc={rc}, n={n}, l={l}, zeta={zeta}, minreg={minreg}:")
            print(f"Error = {minerr} eV/atom")
            print('Time taken: ',time()-start)
            print("---------------------------------------------------------------------------------")
```

# Optimise No. of Samples and Features
Here we create learning curves by varying the number of (PCovFPS-sorted) training samples and features.

## Optimise Number of Samples

```python
#Loop for determining optimal numbers of samples and features...
results=[]

val_feat=get_features(val_frames, batch_size=250) #get features for validation set

#Iterate n_samples
for samples in range(5000, 16000, 1000):
        start = time()
        n_train=samples
        train_feat=get_features(train_frames[:n_train], batch_size=250) #get features for training set

        #Normalise feature vectors
        for feat in train_feat:
            feat/=np.linalg.norm(feat)
        for feat in val_feat:
            feat/=np.linalg.norm(feat)

        K = np.dot(train_feat,train_feat.T)**zeta #compute covariance matrix for training
        Kval = np.dot(val_feat,train_feat.T)**zeta #compute covariance matrix for validation

        #Iterate over regularization values
        minerr = 1e24 ; minreg = 0
        for reg in np.logspace(-9,-5,5):
            KRR = KernelRidge(kernel='precomputed',alpha = reg)
            KRR.fit(K,yred_train[:n_train])
            err = np.std(yred_val - KRR.predict(Kval))
            if err < minerr:
                minerr = err.copy()
                minreg = reg.copy()
        results.append([minerr, n_train, minreg])
        print(f"Results using n_samples={n_train}, minreg={minreg}:")
        print(f"Error = {minerr} eV/atom")
        print('Time taken: ',time()-start)
        print("---------------------------------------------------------------------------------")
```

# Create Model

We will now create our GAP model using Librascal. Most of what happens below should be well-documented, but please message/email me if you have any questions.

The model will be created using the following paramters:
- First 15000 structures from the PCovFPS-sorted _initial_ training set
- The optimal SOAP hyperparameters outlined above
- 2500 PCovFPS-sorted features from the full feature vector
- 2000 sparse points per species
- Regularisers of 0.02 and 2 for energies and forces, respectively
   - This was optimised in loop found in the `delta_general.ipynb` notebook

__Note__: Here we used 2000 sparse points per species, but in one of the initial notebooks `general_purpose_potential_part_x` we optimised the number of sparse points and got the following results:
- H : 1778
- C : 1467
- N : 1467
- O : 1778

Using more sparse points here will give you better results, but 2000 per species is really the upper limit if you would like to maintain some sense of computational efficiency with these hyperparameters.
